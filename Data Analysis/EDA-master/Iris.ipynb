```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/Iris.csv', sep = ',')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.drop(['Id'], axis=1).cov()
```

```python
df['SepalLengthCm'].hist()
plt.show()
```

```python
plt.hist(df['SepalLengthCm'], bins = 10, density = True)
plt.show()
```

```python
species_count = df.groupby('Species')['Id'].count()
species_count
```

```python
plt.bar(species_count.index.values, species_count.values)
plt.show()
```

```python
df.boxplot(column=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'])
plt.show()
```

```python
df.boxplot(column=['SepalLengthCm'], by='Species', )
plt.show()
```

```python
df.boxplot(column=['SepalWidthCm'], by='Species')
plt.show()
```

```python
df.boxplot(column=['PetalLengthCm'], by='Species')
plt.show()
```

```python
df.boxplot(column=['PetalWidthCm'], by='Species')
plt.show()
```

```python
sns.boxplot(x= "Species" , y="PetalLengthCm", data=df)
plt.show()
```

```python
sns.violinplot(x="Species",y="PetalLengthCm",data=df)
plt.show()
```

```python
sns.pairplot(df.drop(['Id'], axis=1), hue="Species")
plt.show()
```
