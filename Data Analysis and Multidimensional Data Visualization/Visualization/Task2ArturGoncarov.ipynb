```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas.plotting import scatter_matrix
from pandas.plotting import andrews_curves


data = pd.read_csv('C:/Users/avgon/Documents/Computer Science/Kompiuterinis Modeliavimas/Multidimensional Data Visualization/go_track_track_full.csv')

data.head()
```

Histogramm and Scatter Plot

```python
data['distance'].hist()
```

```python
#scatter plot
data.plot.scatter(x='time', y='speed', figsize=(8,6), title='Go track data')
```

Scatter Matrix


```python
# Find the columns where each value is null
empty_cols = [col for col in data.columns if data[col].isnull().all()]
# Drop these columns from the dataframe
data.drop(empty_cols,
        axis=1,
        inplace=True)

fig, ax = plt.subplots(figsize=(12,12))
scatter_matrix(data, alpha=1, ax=ax)
```

```python
#check types
data['datetime'] = pd.to_numeric(data['datetime'], errors = 'coerce')
cols = data.columns[data.dtypes.eq('object')]
print(data.dtypes)
data.applymap(np.isreal)
```

Andrew's curves

```python
plt.figure(figsize=(20,10))

andrews_curves(data, 'linha')

```
